Using the data from Shen

Load data:

```python
train_1_0=pd.read_csv('train_1_0_volume.csv')
train_1_1=pd.read_csv('train_1_0_volume.csv')
train_2_0=pd.read_csv('train_2_0_volume.csv')
train_3_0=pd.read_csv('train_3_0_volume.csv')
train_3_1=pd.read_csv('train_3_1_volume.csv')
```

用Train来将数据放起来（好操作）

```python
train=[train_1_0,train_1_1,train_2_0,train_3_0,train_3_1]
```

```python
for i in range(0,len(train)):
    train[i].pop('direction')
    train[i].pop('tollgate_id')
```

`train_in_days`是一个存放不同星期的数据的list，存放第一个数据集的七天，每个item是一天。

```python
train_in_days=[]
for i in range(1,8):
    train_in_days.append(train[0][train[0].weekday==i])
```

将结果plot的出来，看看是否有相似性：

```python
train_in_days[0].head()
```

```python
for i in range(0,len(train_in_days)):
    fig=plt.figure(str(i+1))
    train_in_days[i].volume.plot()
    day="Day "+str(i+1)
    plt.xlabel(day)
    plt.ylabel('Volume')
#     train_in_days[i].volume.plot()
```
